```python
import pandas as pd
job_dataset = pd.read_csv(r"job_descriptions.csv")
display(job_dataset)
```

```python
missing_values = job_dataset.isnull().sum()
display("Missing Values:\n", missing_values)
```

```python
job_dataset = job_dataset.dropna()
display(job_dataset)
```

```python
# Example: Convert 'Job Posting Date' to datetime format
job_dataset['Job Posting Date'] = pd.to_datetime(job_dataset['Job Posting Date'])
```

```python
# covert all data to lower case
job_dataset = job_dataset.applymap(lambda x: x.lower() if isinstance(x, str) else x)
job_dataset.head()
```

```python
import string

def remove_punctuation_from_cell(cell):
    translator = str.maketrans("", "", string.punctuation)
    return str(cell).translate(translator)

for column in job_dataset.columns:
    if  job_dataset[column].dtype == 'object': 
        job_dataset[column] = job_dataset[column].apply(remove_punctuation_from_cell)
        
job_dataset.head()
```

```python
  # adding a standardized format with a hyphen in Salary Range Column
def clean_salary_range(salary_range):
    min_salary = salary_range[:2]
    max_salary = salary_range[3:]
  
    standardized_salary_range = f"{min_salary}-{max_salary}"
    
    return standardized_salary_range

job_dataset['Salary Range'] = job_dataset['Salary Range'].apply(clean_salary_range)
# display(job_dataset)
```

```python
import re
# putting spaces b2n words in Compony profile column
def add_spaces(text):
    clean_text = re.sub(r'(sector|industry|other_keyword)', r'\1 ', text)
    
    return clean_text

job_dataset['Company Profile'] = job_dataset['Company Profile'].apply(add_spaces)

def add_space_before_industry(text):
    return text.replace('industry', ' industry')

job_dataset['Company Profile'] = job_dataset['Company Profile'].apply(add_space_before_industry)
# display(job_dataset)
```

```python
# Have created a version of my dataset, of the cleaned data
job_dataset.to_csv('portalCleaned.csv', index=False)
```

```python
import pandas as pd
tokenizing_frame = pd.read_csv('working_dataset.csv')
display(tokenizing_frame)    
```

```python
from nltk.tokenize import word_tokenize


tokenizing_frame['Qualifications'] = tokenizing_frame['Qualifications'].apply(lambda x: word_tokenize(str(x)))
tokenizing_frame['Work Type'] = tokenizing_frame['Work Type'].apply(lambda x: word_tokenize(str(x)))
tokenizing_frame['Preference'] = tokenizing_frame['Preference'].apply(lambda x: word_tokenize(str(x)))
tokenizing_frame['Job Portal'] = tokenizing_frame['Job Portal'].apply(lambda x: word_tokenize(str(x)))

display(tokenizing_frame)
```

```python
from nltk.tokenize import sent_tokenize

tokenizing_frame['location'] = tokenizing_frame['location'].apply(lambda x: sent_tokenize(x))
tokenizing_frame['Country'] = tokenizing_frame['Country'].apply(lambda x: sent_tokenize(x))
tokenizing_frame['Company'] = tokenizing_frame['Company'].apply(lambda x: sent_tokenize(x))
tokenizing_frame['Role'] = tokenizing_frame['Role'].apply(lambda x: sent_tokenize(x))
tokenizing_frame['Job Title'] = tokenizing_frame['Job Title'].apply(lambda x: sent_tokenize(x))
tokenizing_frame['Job Description'] = tokenizing_frame['Job Description'].apply(lambda x: sent_tokenize(x))


display(tokenizing_frame)
```

```python

# toknizing the skills using the custom spacy
import spacy

nlp = spacy.load("custom_ner_model")

dataset = tokenizing_frame.to_dict(orient="records")

text_column = "skills"

for row in dataset:
   
    text = row[text_column]
    
    doc = nlp(text.lower())  
  
    identified_skills = [ent.text for ent in doc.ents if 
    ent.label_ == "SKILL"]
    
    
    row["tokenized_skills"] = identified_skills

tokenizing_frame = pd.DataFrame(dataset)


display(tokenizing_frame)
```

```python
tokenizing_frame.to_csv('tokenized.csv', index=False)
```
